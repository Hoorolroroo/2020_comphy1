```python
import numpy as np
import matplotlib.pyplot as plt

i,f= 0,10
num_dot = 1000
dt= (f-i)/num_dot

y, v = 1,0
tdata, ydata, vdata=[i], [y], [v]

for idx in range(num_dot+1):
    a=-y
    v +=a*dt
    y +=v*dt+0.5*a*dt**2
    tdata.append(idx*dt+i)
    vdata.append(v)
    ydata.append(y)

plt.figure(figsize=(10,5))
plt.plot(tdata,vdata, label='v')
plt.plot(tdata,ydata, label='y')
plt.legend(fontsize='x-large')
```

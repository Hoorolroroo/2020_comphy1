```python
import csv
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

f=open('c:/python/data/worldwide.csv')
reader = csv.reader(f)
corona = list(reader)

confirmed=[float(x[1]) for x in corona[1:]]
plt.plot(confirmed)
```

```python
diff=[]
diff.append(confirmed[0])
i=0
for i in range(69):
    diff.append(confirmed[i+1]-confirmed[i]) #dx가 어차피 1이기 때문에 생략 하였다.
```

```python
plt.figure(figsize=(15, 7))
plt.plot(diff,label='newly confirmed')
plt.legend(fontsize='large');
plt.xlabel("days after corona virus first occured", fontsize='large')
plt.ylabel("numbers of people", fontsize='large')
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def Leibnizpi(n):
    
    i=0
    pi=0
    a=1
    
    while(i<n):
        b=2*i+1
        pi=pi+a/b
        a=a*-1
        i=i+1
    
    return pi

print(Leibnizpi(50))
```

```python
ydata = []
x=1
for x in range(200):
    ydata.append(Leibnizpi(x+1))
xdata = []
i=1
for i in range(200):
    xdata.append(i+1)
```

```python
plt.figure(figsize=(15, 7)) #n=200 까지로 좀 더 극대화 해보았다. 100부터는 좀 더 미세하게 pi/4로 가까워 지고 있다.
plt.xlabel("n", fontsize='x-large')
plt.ylabel("The magnitude of number", fontsize='x-large')
plt.plot(xdata, ydata, label="Leibniz formula for pi")
y1data=[]
i=1
alpha=np.pi/4
for i in range(200):
    y1data.append(alpha)

plt.plot(xdata, y1data, label="pi/4")
plt.legend(fontsize='x-large');
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize

data=np.loadtxt("c:/python/millikan.txt",float)
xdata=[i[0] for i in data] 
xdata=np.array(xdata)
ydata=[i[1] for i in data]
ydata=np.array(ydata)
plt.xlabel('frequency (Hz)')
plt.ylabel('Voltage (V)')
plt.plot(xdata,ydata,'-o');

def func(x, a, b):
    return a*x+b

popt, pcov = scipy.optimize.curve_fit(func, xdata, ydata)

plt.plot(xdata, func(xdata,popt[0],popt[1]),label='fitted line')
plt.legend();
```

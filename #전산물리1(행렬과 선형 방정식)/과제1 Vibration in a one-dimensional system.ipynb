```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
b=np.zeros(100)
b[0]=1
```

```python
size = 100
d, k = 16, 10
M = np.zeros((size, size))
for i in range(1, size):
    M[i, i - 1] = -k
    M[i, i] = d
    M[i - 1, i] = -k
M[0, 0], M[size - 1, size - 1] = d - k, d - k
```

```python
M
```

```python
a=np.linalg.solve(M,b)
```

```python
plt.figure(figsize=(10,5))
plt.plot(a, '-o',label="a, where C=1")
plt.xlabel("i", fontsize='x-large')
plt.ylabel("Amplitude", fontsize='x-large')
plt.legend(loc='upper right',fontsize='x-large')
```

```python
b_=np.zeros(100)                               
b_[0]=2
a_=np.linalg.solve(M,b_)
plt.figure(figsize=(10,5))
plt.plot(a_, '-o',color='darkorange',label="a, where C=2")
plt.xlabel("i", fontsize='x-large')
plt.ylabel("Amplitude", fontsize='x-large')
plt.legend(loc='upper right',fontsize='x-large')
```

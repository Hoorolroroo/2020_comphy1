```python
from scipy import signal
import numpy as np 
x = np.linspace(0, 10, 500)
plt.plot(x, signal.square(x),label='square wave');

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.ylim(-2, 2)
plt.show()
```

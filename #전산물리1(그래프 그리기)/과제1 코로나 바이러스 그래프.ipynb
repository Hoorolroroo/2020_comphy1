```python
# 다음의 코드를 이용하세요.
import csv

f=open('c:/python/corona virus/covid-19-master/data/countries.csv', newline='')
reader = csv.reader(f)
countries = list(reader)

name=[str(x[1]) for x in countries[1:]]
# n_italy=[[float(x[2]),float(x[3]),float(x[4])] for x in countries[1:] if x[1]=='Italy']
```

```python
nc=[]
ind=0
for i in name:
    temp=[[i,float(x[2]),float(x[3]),float(x[4])] for x in countries[1:] if x[1]==i]
    nc.append(temp)
    ind += 1
```

```python
nc[84]
```

```python
from math import*
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

adayconfirmed= [i[1] for i in nc[84]]
confirmed=np.cumsum(adayconfirmed)
adaydeaths= [i[2] for i in nc[84]]
deaths=np.cumsum(adaydeaths)
```

```python
plt.plot(confirmed, label='cumulative confirmed')
plt.plot(deaths, label='cumulative deaths');
plt.xlabel("days after corona virus first occured")
plt.ylabel("numbers of people")
plt.title("Italy cumulative corona-virus confirmed and deaths")
plt.legend()
```

```python
days = []
x=1
for x in range(70):
    days.append(x+1)
plt.bar(days, adayconfirmed, label='daily confirmed')
plt.bar(days, adaydeaths, label='daily deaths')
plt.xlabel("days after corona virus first occured")
plt.ylabel("numbers of people")
plt.title("Italy daily corona-virus confirmed and deaths")
plt.legend()
```
